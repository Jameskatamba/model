```python
import pandas as pd
import numpy as np
df=pd.read_csv("consumption_data.csv")
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = df.drop(['Water','Date','w2','W','Weight'], axis=1)
y = df['Water']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .75)
X_test
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
RT = DecisionTreeRegressor(random_state=1234);
```

```python
RT.fit(X_train,y_train)
```

```python
#text_representation = tree.export_text(RT)
#print(text_representation)
```

```python
#tr = pd.DataFrame([[12,40,33,1],[12,23,33,6]])
prediction = RT.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mae = mean_absolute_error(y_test, prediction)
mae
```

```python
r2 = r2_score(y_test, prediction)
r2
```

```python
import pickle
pickle.dump(RT, open('model','wb'))
```
